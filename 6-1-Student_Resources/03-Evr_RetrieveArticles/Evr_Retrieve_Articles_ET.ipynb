```python
# Dependencies
import requests
from config import api_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
query = "butterflies"

# Search for articles published between a begin and end date
begin_date = "20140101"
end_date = "20220101"

# Build url

query_url = url  + "&q=" + query + "&begin_date=" + begin_date + "&api-key=" + api_key
```

```python
# Retrieve articles
articles = requests.get(query_url).json()
articles
```
